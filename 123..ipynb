```python
location_review = []
for i in location:
    location_review.append(i.text)
```

```python
def splitting(data):
    average_price = []
    for i in data:
        s = i.split('/')[0]
        average_price.append(s)
    return average_price
```

```python
x= splitting(location_review)
```

```python
x
```

```python
list_1 = []
list_2 = []

for item in x:
    if item.startswith("₹"):
        list_1.append(item)
    elif item.endswith("+"):
        list_2.append(item)
    else:
        list_1.append(item)  # If 'NA', add it to list_1

```
